```python
#Empezamos a trabajar con el csv que contiene los detalles de los avisos existentes (online y offline)
#Vemos la informacion de cada columna

avisosDetalle.info()
```

```python
#Nos quedamos solo con las columnas que creemos que nos pueden servir y eliminamos los campos nulos

avisosDetalle = avisosDetalle[['idaviso','titulo','nombre_zona','tipo_de_trabajo','nivel_laboral','nombre_area','denominacion_empresa']]

avisosDetalle.dropna(inplace=True)
```

```python
#Vemos si todas las publicaciones estan dentro de Buenos Aires (Sino deben filtrarse)

avisosDetalle.groupby('nombre_zona')['idaviso'].count()
```

```python
#Agrupamos todas las publicaciones que corresponden al mismo area y vemos cuales son las que cuentan con mayor cantidad de publicaciones

avisosDetalle.groupby('nombre_area')['idaviso'].count().sort_values(ascending = False).head() 
```

```python
#Y hacemos un grafico de  barras con lo obtenido

avisosDetalle.groupby('nombre_area')['idaviso'].agg('count').sort_values(ascending = False)[0:5].plot(kind='bar', rot=90, figsize=(12,8));

plt.title('Areas con mayor cantidad de publicaciones', fontsize=20);
plt.xlabel('Area', fontsize=16);
plt.ylabel('Cantidad de publicaciones', fontsize=16);
```
